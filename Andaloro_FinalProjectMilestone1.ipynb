```python
import colour
import colour_datasets
import pandas as pd

print(colour_datasets.datasets())
```

```python
dataset = colour_datasets.load('8314702')
dataset.keys()
```

```python
camdata = dataset['Estimated']
camdata
```

```python
df = pd.DataFrame(camdata)
df.shape
```

```python
df.head()
```

```python
colour.plotting.plot_single_cmfs([
    camdata['Olympus E PM2']
],
    bounding_box=(400, 700, 0, .5),
)
```

```python
colour.plotting.plot_single_cmfs([
    camdata['LEICA M (Typ 262)']
],
    bounding_box=(400, 700, 0, .5),                                
)
```

```python
colour.plotting.plot_multi_cmfs([
    camdata['Olympus E PM2'],
    camdata['LEICA M (Typ 262)'],
],
    bounding_box=(400, 700, 0, 0.5),                               
)
```

```python
colour.plotting.diagrams.plot_chromaticity_diagram([
        camdata['Olympus E PM2']
]);
```

```python
colour.plotting.diagrams.plot_chromaticity_diagram([
        camdata['LEICA M (Typ 262)']
]);
```
